```python
import numpy as np
from scipy.stats import poisson
from scipy.stats import binom
from scipy.special import factorial
import matplotlib.pyplot as plt
import math
```

```python
m=1000
lSup = 10
d = np.arange(0,lSup)
mu1 = 1
mu2 = 2
n = 100
p1 = mu1/n
X1 = poisson.pmf(d, mu1)
X2 = poisson.pmf(d, mu2)
Y1 = X1 + X2
Y2 = poisson.pmf(d, mu1+mu2)

plt.plot(X1, color='black', label='X1')
plt.plot(X2, color='blue',  label='X2')

# plt.plot(Y1, color='green', label='Y1')

plt.plot(Y2, color='orange', label='Y2')

c1 = poisson.rvs(mu=mu1, size=m)
c2 = poisson.rvs(mu=mu2, size=m)
h1 = np.histogram(c1, bins=lSup, range=(0,lSup))
h2 = np.histogram(c2, bins=lSup, range=(0,lSup))

Yc = c1 + c2
Yh = np.histogram(Yc, bins=lSup, range=(0,lSup))

plt.plot(h1[1][:-1], h1[0]/m, label='h1', color='purple')
plt.plot(h2[1][:-1], h2[0]/m, label='h2', color='pink')
plt.plot(Yh[1][:-1], Yh[0]/m, label='Yh', color='brown')
plt.legend()

```

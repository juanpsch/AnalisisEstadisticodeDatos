Simular una variable aleatoria Y que sigue una distribucion de
Poisson con parametro μ = 10. Suponer que el resultado de la simulacion es Y = k.
A continuacion simular k variables de Bernoulli Xi con i = 1, ..., k con parametro
p = 0.7. Calcular el valor de una nueva variable Z definida como Z = Pk i=1 Xi.
Repetir la simulacion 1000 veces, construir un histograma de frecuencia de Z y
comparar con una distribucion de Poisson. Nota: usar la distribucion exacta sin
ajustar los datos.

```python
from scipy.stats import poisson
from scipy import stats
import numpy as np
import matplotlib.pyplot as plt
```

```python
p=0.7 ## Probabilidad de éxito de la variable de bernoulli
bernoulli = stats.bernoulli(p)

# Graficando Bernoulli
x = np.arange(-1, 3)
fmp = bernoulli.pmf(x) # Función de Masa de Probabilidad
fig, ax = plt.subplots()
ax.plot(x, fmp, 'bo')
ax.vlines(x, 0, fmp, colors='b', lw=5, alpha=0.5)
ax.set_yticks([0., 0.2, 0.4, 0.6])
plt.title('Distribución Bernoulli')
plt.ylabel('probabilidad')
plt.xlabel('valores')
plt.show()
```

```python
m = 1000 ## iteraciones
mu=10
k = poisson.rvs(mu=mu,size=m) ## Array random de distribución poisson con m elementos

Z = np.zeros(m)
## Array para almacenar resultado de las iteraciones. Iniciado todo en 0

for i in range(m):
    X = bernoulli.rvs(k[i]) # Aleatorio de bernoulli de la iteración en curso (0, 1)
    Z[i] = sum(X) # Contador de éxitos de la iteración en curso
    
bins = 20
lSup = 20

ZHist = np.histogram(Z, bins = lSup, range=(0,lSup))[0]
## Distribución de frecuencias de la suma de todas las variables


muSuma=mu*p
## Nuevo mu para la distribución teórica de la sumatoria de variables de Bernoulli
    
pois1 = poisson.rvs(mu=muSuma, size=m)
# Aleatorios de Poisson con el nuevo mu

poisHist1= np.histogram(pois1, bins = lSup, range=(0,lSup))[0]
## Distribución de frecuencias de pois1

poisHist2 = poisson.pmf(mu=muSuma, k=np.arange(lSup))
## Distribución de frecuencias teórico

ax=plt.plot()
# ax=plt.hist(Z, bins=bins, label='Z', density='True', color='yellow')
ax=plt.plot(ZHist/m, label='Z', color='black')
ax=plt.plot(poisHist1/m, label='Pois1')
ax=plt.plot(poisHist2, label='Pois2')

plt.xlim=(0,20)
plt.legend()
plt.plot()
```
